```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100,1)-0.5
A = np.random.rand(100,100)-0.5

maxiters = 50

(w,v) = np.linalg.eig(A)
plt.figure(); plt.subplot(2,1,1);plt.plot(np.abs(w)); plt.title('Eigenvalues of random matrix'); plt.xlabel('eigenvalue number')
plt.subplot(2,1,2); plt.plot(x); plt.title('Vector x'); plt.xlabel('component')
```

```python
# random matrix multiplication
u = x.copy()
for iters in range(maxiters):
    u = np.dot(A,u)
plt.figure()
plt.plot(u)
```

```python
# random matrix multiplication with relu
u = x.copy()
for iters in range(maxiters):
    u = np.dot(A,u)
    u[np.where(u<0)] = 0 # apply relu
    
plt.figure()
plt.plot(u)
```
